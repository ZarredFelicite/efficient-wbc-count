```python
import tensorflow as tf
import onnx
import keras as keras
import torch
import torch.nn as nn
import torch.nn.functional as F
from HelperFunctions import *
from torch.autograd import Variable
```

```python
device = torch.device(0 if torch.cuda.is_available() else 'cpu')
device_name = torch.cuda.get_device_name(0)
```

```python
model = AlexNet().to(device)
model_file = "Models/AlexNetv2.1s=64.pt"
check_point = torch.load(model_file)
model.load_state_dict(check_point['model_state_dict'])
print(model)
```

```python
# Export the trained model to ONNX
dummy_input = Variable(torch.randn(1, 3, 64, 64).to(device))
torch.onnx.export(model, dummy_input, "Models/AlexNet.onnx", input_names=["input"], output_names=["output"])
```

```python
converter = tf.lite.TFLiteConverter.from_frozen_graph(
        "Models/resnet18/saved_model.pb", ['input'], ['output'])
tflite_model = converter.convert()
open("Models/resnet18.tflite", "wb").write(tflite_model)
```
