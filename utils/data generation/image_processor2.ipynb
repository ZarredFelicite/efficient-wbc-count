```python
#TILE IMAGES
pos = [[0,0],[160,0],[320,0],[480,0],[0,160],[160,160],[320,160],[480,160],[0,320],[160,320],[320,320],[480,320],[0,480],[160,480],[320,480],[480,480]]

def chunks(l, n):
    n = max(1, n)
    return (l[i:i+n] for i in range(0, len(l), n))

def tile_img(input, output, cell, j):
    size = 0.25
    in_path = Path(input)
    save_path = Path(output)
    (save_path / 'images').mkdir(parents=True, exist_ok=True)
    (save_path / 'labels').mkdir(parents=True, exist_ok=True)
    img_list = [img_path.as_posix() for img_path in in_path.rglob('*.png')]
    img_list = chunks(img_list, 16)
    for img_sublist in list(img_list):
        #base image
        bg = Image.new(mode="RGB", size=(640,640))
        save_name = str(j).rjust(6, '0') + '_' + str(cell)
        with open(save_path.as_posix() + '/labels/' + save_name + '.txt', 'w') as f:
            for i in range(len(img_sublist)):
                img = Image.open(img_sublist[i])
                bg.paste(img, box=tuple(pos[i]))
                f.write('{} {} {} {} {}\n'.format(cell,(pos[i][0]+80)/640,(pos[i][1]+80)/640,size,size))
        save_f = save_path.as_posix() + '/images/' + save_name + '.png'
        print(save_f)
        bg.save(save_f)
        j+=1
    return j
```

```python
classes = ['Neutrophil','Lymphocyte','Monocyte','Eosonophil','Basophil']
colour = 'Purple'
j=0
for dset in ['test','valid','train']:
    for cell in [0,1,2,3,4]:
        j = tile_img(input="WBC_Classification_3172_c/{}/{}/{}".format(colour,dset,classes[cell]), output='wbc_det_tile/{}/{}'.format(colour,dset), cell=cell, j=j)
```
